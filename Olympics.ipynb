2. Which country has won the most men's gold medals in Athletics over the years? Sort the results alphabetically by the player's names.

5. Which five countries have won the most medals in recent years (from 2000 to 2008)?

```python
olympics["Edition"].value_counts()
```

```python
# As we can see up there Top edition is 2008 so We gonna pass our condition form 2000 to above 
rec_medals = olympics[(olympics['Edition'] >= 2000)]
rec_medals
```

```python
recent_medals = rec_medals.sort_values(by='Edition', ascending=False)
recent_medals
```

```python
recent_medals['NOC'].value_counts().head(n=5)
```
